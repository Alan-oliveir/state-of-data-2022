# State of Brazil 2022 - Análise Exploratória

## 1\. Introdução

### 1\.1\. Contexto

O State of Data Brazil, realizado pelo Data Hackers e pela Bain & Company, é uma das maiores pesquisas sobre o panorama do mercado de trabalho brasileiro na área de dados.  
Dentre os assuntos mapeados na última pesquisa, realizada entre 10 de outubro de 2022 e 28 de novembro de 2022, estão o perfil atual das três maiores pprofissões da área de dados do Brasil: analista de dados, cientista de dados e engenheiro de dados.  
Veja mais detalhes no site da pesquisa: www.stateofdata.com.br


### 1\.2\. Objetivo

Analisar a distribuição das faixas salariais para os profissionais de nível júnior para o cargo de analista, cientista e engenheiro de dados.

## 2\. Dados

### 2\.1\. Descrição

Os dados da pesquisa foram disponibilizados no Kaggle ([link](https://www.kaggle.com/datasets/datahackers/state-of-data-2022)). O arquivo está no formato **CSV**, possui **353 colunas** e **4271 linhas** tem aproximadamente **10 MB**.

## 3\. Processamento

### 3\.1\. leitura

```python
import pandas as pd

data = list()

with open(file='State_of_data_2022.csv', mode="r") as fp:
  header = fp.readline()
  line = fp.readline() # linhas como texto

  while line:
    line = line.split(sep=',') # linha como lista
    data.append({"Cargo": line[21], "Experiência": line[22], "Salário": line[23]})
    line = fp.readline() # linhas como texto

data_table = pd.DataFrame(data)
```

```python
data_table.head(20)
```

```python
data_table.shape
```

### 3\.2\. Limpeza

#### 3\.2\.1\. Descrição

**Objetivo**: uma boa análise só pode ser obtida com dados consistentes, portanto a limpeza de dados se faz necessária.

**Descrição**: remoção de linhas com dados vazios, incorretas e sem relação com os cargos desejados.

- Para esta análise iremos remover:
 - Linhas vazias
 - Linhas corrompidas

- Mantemos apenas as linhas nos quais os cargos são:
 - Analista de dados
 - Cientista de dados
 - Engenheiro de dados

#### 3\.2\.2\. Linhas vazias

```python
data_table = data_table.replace('', pd.NA)
data_table = data_table.dropna()
```

```python
data_table.head(20)
```

```python
data_table.shape
```

#### 3.2.3. Linhas corrompidas

```python
data_table['Salário'] = data_table['Salário'].apply(lambda row: row if 'R$' in row else pd.NA)
data_table = data_table.dropna()
```

```python
data_table.head(20)
```

```python
data_table.shape
```

#### 3.2.4. Cargos de interesse

```python
data_table['Cargo'].unique()
```

```python
positions = ['Analista de Dados/Data Analyst', 'Cientista de Dados/Data Scientist', 'Engenheiro de Dados/Arquiteto de Dados/Data Engineer/Data Architect']
data_table['Cargo'] = data_table['Cargo'].apply(lambda row: row if row in positions else pd.NA)
data_table = data_table.dropna()
```

```python
data_table.head(20)
```

```python
data_table.shape
```

- Para facilitar a análise alteramos os nomes dos cargos

```python
labels = ['Analista', 'Cientista', 'Engenheiro']
labels = dict(zip(positions, labels))
```

```python
print(labels)
```

```python
data_table['Cargo'] = data_table['Cargo'].apply(lambda row: labels[row])
```

```python
data_table.head(n=20)
```

- Com os dados limpos e nomes alterados, salvamos os dados em um novo arquivo CSV.

```python
data_table.to_csv("data.csv", header=True, index=False)
```

## 4\. Insights

### 4\.1\. Análise da quantidade de pessoas do nível júnior por faixa de salário e cargo.

**Agrupamento e Agregações**

Cria-se uma coluna unitária chamada Quantidade para facilitar o agrupamento e agregações.

```python
data_table['Quantidade'] = 1
```

```python
data_table.head(n=20)
```

Agrega-se o dado de interesse, aplicando a operação de soma na coluna Quantidade recém criada.

```python
salario_cargo_experiencia_agg = data_table[['Cargo', 'Experiência', 'Salário', 'Quantidade']].groupby(['Cargo', 'Experiência', 'Salário']).agg('sum').reset_index()
```

```python
salario_cargo_experiencia_agg.head(20)
```

Após fazer a agregação vamos ordenar os dados pelo salário.

```python
salario_ordem = {
    'Menos de R$ 1.000/mês': 0,
    'de R$ 1.001/mês a R$ 2.000/mês': 1,
    'de R$ 2.001/mês a R$ 3.000/mês': 2,
    'de R$ 3.001/mês a R$ 4.000/mês': 3,
    'de R$ 4.001/mês a R$ 6.000/mês': 4,
    'de R$ 6.001/mês a R$ 8.000/mês': 5,
    'de R$ 8.001/mês a R$ 12.000/mês': 6,
    'de R$ 12.001/mês a R$ 16.000/mês': 7,
    'de R$ 16.001/mês a R$ 20.000/mês': 8,
    'de R$ 20.001/mês a R$ 25.000/mês': 9,
    'de R$ 25.001/mês a R$ 30.000/mês': 10,
    'de R$ 30.001/mês a R$ 40.000/mês': 11,
    'Acima de R$ 40.001/mês': 12
}

salario_cargo_experiencia_agg['Ordem'] = salario_cargo_experiencia_agg['Salário'].apply(lambda row: salario_ordem[row])
salario_cargo_experiencia_agg = salario_cargo_experiencia_agg.sort_values(by=['Cargo', 'Experiência', 'Ordem'], ascending=[True, True, True])
```

```python
salario_cargo_experiencia_agg.head(20)
```

**Visualização**

Por fim, vamos filtrar os pprofissionais de nível de experiência júnior e gerar o gráfico de barras.

```python
data = salario_cargo_experiencia_agg[salario_cargo_experiencia_agg['Experiência'] == 'Júnior']
```

```python
import seaborn as sns

with sns.axes_style('whitegrid'):
  plot = sns.barplot(data=data, x='Cargo', y='Quantidade', hue='Salário', palette='rocket_r')
  plot.set_title('Faixas de salário por cargo para profissionais de nível júnior')
  plot.set_ylabel('Quantidade de profissionais')
  plot.set_xlabel('Cargo')
  plot.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

**Resultados**

- Observa-se que a distribuição dos salários para o cargo de analista se concentra principalmente na faixa de mil a seis mil reais por mês;
- Em contraste, a distribuição de salários para os cargos de cientista e engenheiro de dados é mais concentrada na faixa de quatro a oito mil reais por mês;
- Observa-se ainda que os analistas são a maioria dos profissionais pesquisados.

Conclui-se, portanto, que os profissionais de análise tendem a receber salários inferiores aos profissionais de ciência ou engenharia de dados. No entanto, é importante ressaltar que a quantidade de vagas disponíveis na área pode ser maior para analistas.
